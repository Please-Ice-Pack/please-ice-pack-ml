```python
pt_path = '/content/drive/MyDrive/마켓컬리_해커톤/weight/best.pt'
model =torch.hub.load('./yolov5', 'custom', path=pt_path, source='local')
model.conf = 0.5
```

```python
import json
import time
source = '/content/drive/MyDrive/마켓컬리_해커톤/sample/10_4.jpg'

all_time = 0
for _ in range(10):
  pre = time.time()
  results = model(source)
  detect_res = results.pandas().xyxy[0].to_json(orient="records")
  detect_res = json.loads(detect_res)
  all_time += time.time()-pre
print(all_time / 10)
print(detect_res)

# cpu 기준 1.2 ~ 1.3초 소요
```
